## Web Scrapping Data using bs4 and pandas

```python
from bs4 import BeautifulSoup
import requests
```

#### Webpage url

```python
url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"

page = requests.get(url)
```

```python
page
```

```python
##Beautiful Soup is a Python library for pulling data out of HTML and XML files.
#When you parse an HTML page using BeautifulSoup, 
#it creates this object, allowing to navigate and extract information from the document.
```

```python
soup = BeautifulSoup(page.text, 'html')
#html.parser
#soup = BeautifulSoup(page.text, 'html.parser')

print(soup)
```

#### Get table from the soup (only first table returned)

```python
soup.find('table')
# only return the first table html
```

#### Get all the tables from the soup

```python
# return all the objects with class table from the html page
soup.find_all('table')
# the returned result is a list
```

#### Return the table with index 1 (2nd table)

```python
# return html for the first table
table = soup.find_all('table')[1]

print(table)
```

#### Fetch header for the table (retun header of all tables)

```python
# The <th> tag in HTML defines a header cell in an HTML table.
soup.find_all('th')
# retun all the headers of table from the html webpage
```

#### Fetch rows from all tables

```python
# The <tr> tag in HTML is used to define rows within a table.
soup.find_all('tr')
# return all row records from a table within the html webpage
```

```python
# getting all the titles for all the tables
all_table_titles = soup.find_all('th')

all_table_titles
```

#### Fetch title for table1 only

```python
# getting title for the first table
table_one_title = table.find_all('th')

table_one_title
```

#### Create a df with the table headers (data not inserted)

```python
# writing a for loop to get the title data in a list
header_one = []
for title in table_one_title:
    a = title.text.strip()
    header_one.append(a)
```

```python
header_one
```

```python
# the above for loop can be implemented using list comprehension
header_one = [title.text.strip() for title in table_one_title]
print(header_one)
```

```python
import pandas as pd
```

```python
# creating a dataframe only with headers
df = pd.DataFrame(columns = header_one)
df
```

#### Rows from the first table(using 'tr' tag)

```python
# records for the first table
column_data = table.find_all('tr')
column_data
```

```python
type(column_data)
```

#### Retrive each row as a separate list

```python
# The <td> tag in HTML stands for table data.
# loop through the above 'column_data' in search of <td> tag
for row in column_data[1:]:
    row_data = row.find_all('td')
    records = [row.text.strip() for row in row_data] 
    print(records)
    
# but here the individual results are printed as seperate rows
# but this records are supposed to insert into the df create above
# in the below cell, modify the above for loop to append the data as a list
```

#### Create a list of records (rows) for df

```python
# append the individual results as a list
data = []
for row in column_data[1:]:
    row_data = row.find_all('td')
    records = [row.text.strip() for row in row_data] 
    data.append(records)
    
print(data)
```

#### Creating the complete dataframe

```python
df = pd.DataFrame(data,columns = header_one)
df
```

#### Another method to append each row using df.loc

```python
# another method within the for loop itself appending the data into the dataframe
# will create a second dataframe for the same
df_one = pd.DataFrame(columns = header_one)
df_one
```

```python
len(df_one)
```

```python
# within the for loop make use of the length of the dataframe and insert data
for row in column_data[1:]:
    row_data = row.find_all('td')
    records = [row.text.strip() for row in row_data]
    df_one.loc[len(df_one)] = records
    
df_one      
```
